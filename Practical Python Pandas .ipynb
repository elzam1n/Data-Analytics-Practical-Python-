```python
import pandas as pd
import numpy as np
```

```python
url = "https://en.wikipedia.org/wiki/List_of_countries_by_population_(United_Nations)"

df = pd.read_html(url)[0]
```

```python
df.head()
```

```python
#first row shows total value so we have to drop it
df.drop(0,inplace=True)
df.head()
```

```python
#population change described as percentage but that doesnt make correct insight about increment of population
df['change_in_count'] = df['Population (1 July 2023)']-df['Population (1 July 2022)']
df.head()
```

```python
#now we can see top 10 countries with population increment
df.sort_values(by='change_in_count',ascending=False).head(10)
```

```python
#lets see bottom 10
df.sort_values(by='change_in_count',ascending=True).head(10)
```

```python
#grouping categories 
df.groupby('UN Continental Region[1]')[['Population (1 July 2023)','change_in_count']].sum()
# as we can see Region Europe has huge demographic problem.This explains refugee policy by european countries
```

```python
df.groupby('UN Statistical Subregion[1]')[['Population (1 July 2023)','change_in_count']].sum().sort_values(by='change_in_count')
```
